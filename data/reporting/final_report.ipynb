# Model comparison

In this notebook we will take a look to the final metrics obtained from every dataset and every model trained. We will show some conclusions about accuracy and running times.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
colnames1 = ['Run Time', 'Train AUC', 'Test AUC']
colnames2 = ['Run Time', 'Train RMSE', 'Test RMSE']
```

### Breast cancer

```python
metrics1 = pd.read_csv('../model_output/metrics/breast_cancer/final_metrics.txt', header=None, delimiter=';', index_col=0, names=colnames1)
metrics1.sort_values(by='Test AUC', ascending=False)
```

```python
plt.figure(figsize=(18, 6))

plt.bar(np.arange(len(metrics1.index.values))-0.15, metrics1['Train AUC'], width=0.3, label='Train', edgecolor='k')
plt.bar(np.arange(len(metrics1.index.values))+0.15, metrics1['Test AUC'], width=0.3, label='Test', edgecolor='k')

plt.xlabel('Models', fontsize='x-large')
plt.ylabel('AUC', fontsize='x-large')
plt.title('Accuracy comparison', fontsize='x-large')
plt.legend(loc='lower left', fontsize='large')
plt.xticks(np.arange(len(metrics1.index.values)), metrics1.index.values, fontsize='large')
plt.gca().get_xticklabels()[metrics1.index.values.tolist().index(metrics1['Test AUC'].idxmax())].set_fontweight('bold')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.plot(np.arange(len(metrics1.index.values)), metrics1['Run Time'], 'o-')

plt.xlabel('Models', fontsize='x-large')
plt.ylabel('Time (seconds)', fontsize='x-large')
plt.title('Run times comparison', fontsize='x-large')
plt.xticks(np.arange(len(metrics1.index.values)), metrics1.index.values, fontsize='large')
plt.gca().get_xticklabels()[metrics1.index.values.tolist().index(metrics1['Run Time'].idxmin())].set_fontweight('bold')

plt.show()
```

```python
df_names = ['Cncer de mama', 'Seguros de coche', 'Transformador de electricidad', 'Precios de casas','Defectos de software', 'Superconductores']

plt.figure(figsize=(16, 6))

for model in metrics1.index:
    ranking = []
    for df in [metrics1, metrics2, metrics3, metrics4, metrics5, metrics6]:
        if df is metrics1 or df is metrics2 or df is metrics5:  
            ranking.append(list(df.sort_values(by='Test AUC', ascending=False).index).index(model)+1)
        else:
            ranking.append(list(df.sort_values(by='Test RMSE').index).index(model)+1)
    plt.plot(df_names, ranking, marker='o', label=model, markersize=8)
    
plt.legend(loc='best', fontsize='large',  bbox_to_anchor=(1, 0.75))
plt.ylabel('Ranking', fontsize='x-large')
plt.xlabel('Conjuntos de datos', fontsize='x-large')
plt.title('Comparacin de rendimiento entre modelos', fontsize='x-large')
plt.gca().invert_yaxis()
plt.savefig('rankingmodelos.pdf')
plt.show()
```
